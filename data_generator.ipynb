```python
from faker import Faker
import random
import csv

# Initialize Faker
fake = Faker()
lab_names = [
    "Biology Lab",
    "Chemistry Lab",
    "Physics Lab",
    "Computer Lab",
    "Engineering Lab",
    "Psychology Lab",
    "Geology Lab",
    "Botany Lab",
    "Zoology Lab",
    "Microbiology Lab",
    "Anatomy Lab",
    "Ecology Lab",
    "Neuroscience Lab",
    "Genetics Lab",
    "Biochemistry Lab"
]

# Function to generate fake data for Lab table
def generate_lab(num_records):
    with open('lab.csv', 'w', newline='') as csvfile:
        fieldnames = ['Lab Name', 'Room no', 'Block no', 'Amount Allocated', 'Working Hours', 'Capacity']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for i in range(0,len(lab_names)):
            lab_name = lab_names[i]
            room_no = random.randint(100, 999)
            block_no = random.randint(1, 10)
            amount_allocated = round(random.uniform(1000, 10000), 2)
            working_hours = random.randint(8, 24)
            capacity = random.randint(10, 100)

            writer.writerow({
                'Lab Name': lab_name,
                'Room no': room_no,
                'Block no': block_no,
                'Amount Allocated': amount_allocated,
                'Working Hours': working_hours,
                'Capacity': capacity
            })

# Function to generate fake data for Staff table
def generate_staff(num_records):
    with open('staff.csv', 'w', newline='') as csvfile:
        fieldnames = ['Employee ID', 'First Name', 'Middle Name', 'Last Name', 'Salary', 'Role']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        employee_id_generated = set()
        
        for i in range(num_records):
            employee_id = random.randint(10**9, (10**10 - 1))
            while employee_id in employee_id_generated:
                employee_id= random.randint(10000000, 99999999)
            employee_id_generated.add(employee_id)
            
            first_name = fake.first_name()
            middle_name = fake.first_name() if random.choice([True, False]) else ''
            last_name = fake.last_name()
            salary = round(random.uniform(20000, 100000), 2)
            role = random.choice(['Researcher', 'Technician', 'Assistant'])

            writer.writerow({
                'Employee ID': employee_id,
                'First Name': first_name,
                'Middle Name': middle_name,
                'Last Name': last_name,
                'Salary': salary,
                'Role': role
            })

dept=["CSE","ME","MSE","CL","EE","CE"]
# Function to generate fake data for Department table
def generate_department(num_records):
    with open('department.csv', 'w', newline='') as csvfile:
        fieldnames = ['Dept Name', 'No of Faculties']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for i in range(0,len(dept)):
            dept_name = dept[i]
            no_of_faculties = random.randint(5, 50)

            writer.writerow({
                'Dept Name': dept_name,
                'No of Faculties': no_of_faculties
            })

# Function to generate fake data for Professor table
def generate_professor(num_records):
    with open('professor.csv', 'w', newline='') as csvfile:
        fieldnames = ['Employee ID', 'First Name', 'Middle Name', 'Last Name', 'Grant Aid', 'Office Number']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        employee_id_generated = set()
        
        for i in range(num_records):
            employee_id = random.randint(10**9, (10**10 - 1))
            while employee_id in employee_id_generated:
                employee_id= random.randint(10000000, 99999999)
            employee_id_generated.add(employee_id)
            
            first_name = fake.first_name()
            middle_name = fake.first_name() if random.choice([True, False]) else ''
            last_name = fake.last_name()
            grant_aid = round(random.uniform(0, 100000), 2)
            office_number = fake.building_number()

            writer.writerow({
                'Employee ID': employee_id,
                'First Name': first_name,
                'Middle Name': middle_name,
                'Last Name': last_name,
                'Grant Aid': grant_aid,
                'Office Number': office_number
            })

# Generate fake data for Lab, Staff, Department, and Professor tables
generate_lab(20)
generate_staff(50)
generate_department(10)
generate_professor(30)
```
